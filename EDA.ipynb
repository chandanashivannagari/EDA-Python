```python
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
data=pd.read_csv('Student.csv')
```

```python
print(data)
print()
print(data.describe())
```

```python
#to display unique values in a particular column 
data['Chemistry'].nunique()
```

```python
#nunique displays number of unique values 
#unique displays what are the unique values in it 
data['Result'].unique()
```

Cleaning the data
===

```python
#check for null values
data.isnull().sum()
```

Relationship analysis
===

```python
#cr-relation matrix: summerises data 
corelation=data.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
sns.pairplot(data)
```

```python
sns.relplot(x='Maths',y='Physics',hue='Result',data=data)
```

```python
sns.relplot(x='Maths',y='Physics',hue='Chemistry',data=data)
```

```python
#Histogram
sns.distplot(data['Maths'])

```

```python
sns.histplot(data['Maths'])
```

```python
sns.catplot(x='Maths',kind='box',data=data)
```
